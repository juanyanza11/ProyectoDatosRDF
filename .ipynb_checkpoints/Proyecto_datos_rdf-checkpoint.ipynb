```python
import pandas as pd

from rdflib import URIRef, BNode, Literal
from rdflib import Namespace
from rdflib import Graph, URIRef, Literal, BNode
```

```python
data_jugadores_fifa = pd.read_csv("fifa20_data.csv")
data_jugadores_fifa = data_jugadores_fifa.astype({"Club": str})
data_jugadores_fifa
```

### Prefijos FOAF, RDF

```python
# import prefixes of models that I will reuse:
from rdflib.namespace import FOAF, RDF

# Print some properties:
print(RDF.type)
# == "http://www.w3.org/1999/02/22-rdf-syntax-ns#type"
# == rdflib.term.URIRef("http://xmlns.com/foaf/0.1/knows")
```

### Prefijos vocab, dbo

```python
# prefix for my data
strprefix = "http://vocab.org/"
vocab = Namespace(strprefix) 
vocab

#prefix for external prefix that is not available in rdflib.namespace:
dbo = Namespace('http://dbpedia.org/ontology/')
dbp = Namespace('http://dbpedia.org/property/')
dbr = Namespace('http://dbpedia.org/resource/')
```

```python
print(dbp)
```

```python
# Create my graph:

g = Graph()
```

```python
from unicodedata import normalize
```

```python
#A partir del nombre se fijar√° la URI, por tanto, se necesita formatear el texto para quitar tildes:
a,b = '√°√©√≠√≥√∫√º','aeiouu'
trans = str.maketrans(a,b)

# Recorrer la data e ir creando los datos:
for index, row in data_jugadores_fifa.iterrows():
    soccerPlayersURI = URIRef(dbr + row['Full_Name'].replace(' ', '_').replace('.', '').translate(trans))
    name = Literal(row['Full_Name'])
    age = Literal(row['Age'])
    weight = row['Weight']
    height = row['Height']
    countryURI = URIRef(dbr + row["Country"].replace(' ', '').translate(trans))
    image = URIRef(row['Image'])
    market_value = Literal(row['Value'])
    wage = Literal(row['Wage'])
    foot = Literal(row['foot'])
    overall = Literal(row['Overall'])
    SportClubsURI = URIRef(dbr+ row['Club'].replace(' ', '_').replace('.', '').replace('nan', '').translate(trans))
    label_SportCLubs = row['Club'].replace('nan', '')
    
    cadena_height = height.strip('"').split("'")
    #Transformaci√≥n altura de pies y pulgadas a cm
    pies = int(cadena_height[0])
    pulgadas = int(cadena_height[1])
    
    cm = round(pies * 30.48)
    cm += round(pulgadas * 2.54)

    height_cm = Literal(cm)
    
    #Transformaci√≥n peso en lbs a Kg
    weight_kg = round(int(weight.strip("lbs")) / 2.2)
    weight_kg = Literal(weight_kg)
    
    g.add((soccerPlayersURI, RDF.type, FOAF.Person))
    g.add((soccerPlayersURI, FOAF.name, name))
    g.add((soccerPlayersURI, FOAF.age, age))
    g.add((soccerPlayersURI, FOAF.img, image))
    g.add((soccerPlayersURI, dbo.birthPlace, countryURI))
    g.add((soccerPlayersURI, dbo.weight, weight_kg))
    g.add((soccerPlayersURI, dbo.height, height_cm))
    g.add((soccerPlayersURI, vocab.marketValue, market_value))
    g.add((soccerPlayersURI, vocab.wage, wage))
    g.add((soccerPlayersURI, vocab.overall, overall))
    g.add((SportClubsURI, dbo.currentMember, soccerPlayersURI))
    
```

```python
# Grafo para m√°s de 19000 de jugadores de la FIFA
g.print()
```

```python
g.serialize(destination="fifa-players2020.ttl") # turtle format as a default.
```
